Pensaremos más bien (como debe ser) en los átomos como sistemas que pueden estar o en el estado 1 ("radioactivo") o el estado 0 ("decaído"), y que obedecen un proceso estocástico que gobierna su decaimiento. [1] Tratemos un átomo como sigue. En cada paso, tiene una probabilidad $p$ de decaer del estado $1$ al estado 0.

(i) Escribe una función que calcule el tiempo *hasta que* decaiga.

```julia
function tiempo_decaimiento(p)
    tiempo=0.0
    while rand()>=p
            tiempo+=1.0
    end
    tiempo
end
```

(ii) Encuentra la distribución de tiempos de decaimiento. ¿Cómo es la cola de la distribución? ¿Qué tipo de distribución es? 

```julia
function distribucion(p,t_max=1000000)
    dist=zeros(t_max)
    for i in 1:t_max
        dist[i]=tiempo_decaimiento(p)
    end
    return dist
end
```

```julia
a=distribucion(0.001);
```

```julia
function histograma(X::Array,divisiones=1000)    
    Xmin=minimum(X)
    Xmax=maximum(X)
    intervalo=Xmax-Xmin
    paso=intervalo/divisiones
    cajas=[Xmin:paso:Xmax]
    hist=zeros(length(cajas))
    for x in X
        for i in 1:divisiones
            if cajas[i]<=x && x<cajas[i+1]
                hist[i]+=1
            end
        end
    end
    return cajas,hist/(sum(hist))
end
```

```julia
@time tiempos,cuentas=histograma(a)
```

```julia
plt.subplot(211)
plt.plot(tiempos,cuentas)
plt.title("Distribución de probabilidad de tiempos de decaimiento")

plt.subplot(212)
plt.semilogy(tiempos,cuentas)
plt.title("Distribución de tiempos de decaimiento semilog")
```

#### Con esto podemos ver que se trata de una distribución exponencial, al menos hasta cierto tiempo.

[2] Simula $N$ tales átomos, cada uno de los cuales tiene una probabilidad $p$ de decaer en cada paso.

Dibuja el número de átomos que *no* han decaído todavía hasta el paso $n$, en función de $n$. ¿Qué forma tiene la gráfica? ¿Es sorprendente?

```julia
function N_atomos(N,p,t_max=5000)
    atomos_tiempo=zeros(t_max)
    for tiempo in 1:t_max
        for atomo_n in 1:N
            if tiempo_decaimiento(p)>tiempo
                atomos_tiempo[tiempo]+=1
            end
        end
    end
    return atomos_tiempo
end
```

```julia
@time atomos=N_atomos(1000,0.001);
```

```julia
plt.subplot(211)
plot(atomos)
plt.xlabel("Tiempo")
plt.ylabel("Número de átomos")

plt.subplot(212)
plt.semilogy(atomos)
#plt.title("Semilog en y del número de átomos que no han decaído hasta el tiempo t")
plt.xlabel("Tiempo")
plt.ylabel("log(Número de átomos)")
```

[3] (i) Si $N$ es muy grande, escribe una ecuación aproximada para cómo cambia el número *promedio* de átomos que no se hayan decaído al tiempo $1$, tiempo $2$, ..., tiempo $n$. 

(ii) Resuelve la ecuación para encontrar la proporción no decaída al tiempo $n$.

Para el tiempo $t=1$, nuestro cambio en el promedio de átomos que no han decaído se puede tomar simplemente como $N - pN$ pues cada uno de los átomos tiene probabilidad $p$ de decaer, $i.e.$
\begin{equation}
\langle N(t=1) \rangle = N(1-p)
\end{equation}

\begin{equation}
\Rightarrow \Delta \langle N(t=1) \rangle = N(t=1)-N= -pN
\end{equation}

De tal manera que al tiempo $t=2$ dependerá del número de átomos que había en el tiempo $t=1$, $i.e.$

\begin{equation}
\langle N(t=2) \rangle = \langle N(t=1) \rangle(1-p) = N(1-p)^2
\end{equation}

Y por lo tanto,

\begin{equation}
\Delta \langle N(t=2) \rangle =\langle N(t=2) \rangle -N = N((1-p)^2-1)
\end{equation}

Entonces, para el tiempo $t=n$,

\begin{equation}
\langle N(t=n) \rangle = N(1-p)^n
\end{equation}

\begin{equation}
\Delta \langle N(t=n) \rangle = N((1-p)^n-1)
\end{equation}

Y finalmente la proporción de átomos no decaídos al tiempo $t=n$ será:

\begin{equation}
\frac{\langle N(t=n) \rangle}{N} = (1-p)^n
\end{equation}

Reescribe el resultado de la pregunta 3 en términos de $\delta t$, y toma el límite cuando $\delta t \to \infty$. [Pista: hay que reconocer que aparecerá cierta función interesante en este límite.]


Ahora, si los pasos de tiempo son $\delta t$, podemos expresar la segunda ecuación del inciso anterior como
\begin{equation}
\frac{ \Delta \langle N(\delta t) \rangle}{\delta t} = -pN
\end{equation}

que al tomar el límite cuando $\delta t \to 0$
\begin{equation}
\frac{ dN}{dt} = -pN
\end{equation}

que al resolver se obtiene 
\begin{equation}
N(t)=Ne^{-pt}
\end{equation}
En cuanto al límite cuando $\delta t \to \infty$, tenemos

\begin{equation}
-pN=0
\end{equation}
i.e. $p=0$, lógico, puesto a que la probabilidad de decaimiento  decrece cuando $\delta t$ crece. Así, la ecuación de la población quedará constante
\begin{equation}
N(t)=N
\end{equation}

[5] Da una interpretación de la *tasa de decaimiento*.  ¿Puedes derivar la ecuación diferencial estándar de decaimiento radioactivo de este análisis?

La tasa de decaimiento es una medida de la rapidez del cambio (negativo) de la población de una muestra. Teniendo que para un tipo de átomo existe una probabilidad $p$ de decaer en un instante, si se contara con una muestra o colección de ellos, i.e. una $N$ lo suficientemente grande, tendriamos que $N p$ átomos decayeron, en otras palabras, que habrá decaído una fracción de la población igual a la probabilidad de decaimiento. Por lo que la tasa de decaimiento es justamente la probabilidad de decaimiento de un átomo solo.

\begin{equation}
Np=-\frac{d N}{d t}
\end{equation}
ecuación que resolvemos integrando
\begin{equation}
\int_0^{t'} -p dt=\int_0^{t'}\frac{1}{N}\frac{d N}{d t}dt=\int_N^{N(t')}\frac{1}{N}dN
\end{equation}
\begin{equation}
-pt'=\ln \frac{N(t')}{N}
\end{equation}
lo que finalmente nos lleva la expresión
\begin{equation}
N(t)=Ne^{-pt}
\end{equation}
