En esta tarea, simularemos el decaimiento de $N$ átomos radioactivos. El enfoque usual es en términos de una ecuación diferencial: al parecer esto es evidente para todos, excepto para mí... 

Pensaremos más bien (como debe ser) en los átomos como sistemas que pueden estar o en el estado 1 ("radioactivo") o el estado 0 ("decaído"), y que obedecen un proceso estocástico que gobierna su decaimiento. [El análisis funcionará también, por supuesto, para cualquier otro sistema que se puede modelar así.]

[1] Tratemos un átomo como sigue. En cada paso, tiene una probabilidad $p$ de decaer del estado $1$ al estado 0.

(i) Escribe una función que calcule el tiempo *hasta que* decaiga.

```julia
function tiempo_decaimiento(p)
    tiempo=0.0
    while rand()>=p
            tiempo+=1.0
    end
    tiempo
end
```

(ii) Encuentra la distribución de tiempos de decaimiento. ¿Cómo es la cola de la distribución? ¿Qué tipo de distribución es? 

```julia
function distribucion(p,t_max=1000000)
    dist=zeros(t_max)
    for i in 1:t_max
        dist[i]=tiempo_decaimiento(p)
    end
    return dist
end
```

```julia
a=distribucion(0.001);
```

```julia
function histograma(X::Array,divisiones=1000)    
    Xmin=minimum(X)
    Xmax=maximum(X)
    intervalo=Xmax-Xmin
    paso=intervalo/divisiones
    cajas=[Xmin:paso:Xmax]
    hist=zeros(length(cajas))
    for x in X
        for i in 1:divisiones
            if cajas[i]<=x && x<cajas[i+1]
                hist[i]+=1
            end
        end
    end
    return cajas,hist/(sum(hist))
end
```

```julia
@time tiempos,cuentas=histograma(a)
```

```julia
plt.subplot(211)
plt.plot(tiempos,cuentas)
plt.title("Distribución de probabilidad de tiempos de decaimiento")

plt.subplot(212)
plt.semilogy(tiempos,cuentas)
plt.title("Distribución de tiempos de decaimiento semilog")
```

#### Con esto podemos ver que se trata de una distribución exponencial, al menos hasta cierto tiempo.

[2] Simula $N$ tales átomos, cada uno de los cuales tiene una probabilidad $p$ de decaer en cada paso.

Dibuja el número de átomos que *no* han decaído todavía hasta el paso $n$, en función de $n$. ¿Qué forma tiene la gráfica? ¿Es sorprendente?

```julia
function N_atomos(N,p,t_max=10000)
    atomos_tiempo=zeros(t_max)
    for tiempo in 1:t_max
        for atomo_n in 1:N
            if tiempo_decaimiento(p)>tiempo
                atomos_tiempo[tiempo]+=1
            end
        end
    end
    return atomos_tiempo
end
```

```julia
@time atomos=N_atomos(1000,0.001);
```

```julia
plt.subplot(211)
plot(atomos)
plt.xlabel("Tiempo")
plt.ylabel("Número de átomos")

plt.subplot(212)
plt.semilogy(atomos)
#plt.title("Semilog en y del número de átomos que no han decaído hasta el tiempo t")
plt.xlabel("Tiempo")
plt.ylabel("log(Número de átomos)")
```

Teniendo que para un tipo de átomo existe una probabilidad $p$ de decaer en un cierto tiempo $\delta t$, si se contara con una muestra o colección de ellos, una $N$ lo suficientemente grande, tendriamos que $N\cdot p$ átomos decayeron en un tiempo $\delta t$, i.e. que una fracción de la población igual a la probabilidad de decaimiento.
\begin{equation}
Np=-\frac{\delta N}{\delta t}
\end{equation}

que tomando el límite cuando $\delta t \to 0$, nos lleva a 

\begin{equation}
Np=-\frac{dN}{dt}
\end{equation}

cuya solución lleva a 
\begin{equation}
N(t)=N_oe^{-pt}
\end{equation}

Cuando tomamos el límite contrario, i.e., $\delta t \to \infty$, tenemos que la ecuaión se convierte en
\begin{equation}
Np=0
\end{equation}

es decir, que $p=0$, y al ser cero, la ecuación de la [3] queda

\begin{equation}
\Delta \langle N(\delta t) \rangle =0
\end{equation}
