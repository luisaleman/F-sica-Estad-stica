$P_{t+1}(i)=P_{t}(i)+[pP_{t}(i-1)+qP_{t}(i+1)]-[pP_{t}(i)+qP_{t}(i)]$

Donde el primer corchete es la probabilidad de que salten de los lugares adyacentes a él, y el segundo corchete es la probabilidad de que estando en el lugar, salte a los adyacentes. Haciendo un poco de álgebra,utilizando que $p+q=1$, se puede llegar a:

$P_{t+1}(i)=pP_{t}(i-1)+qP_{t}(i+1)$

```julia
time=1000
middle=time+1
Probabilities=zeros(time,2time+1)
Probabilities[1,middle]=1
for t in 1:(time-1)
    for i in -t:t
        Probabilities[t+1,middle+i]=0.5*(Probabilities[t,middle+i+1]+Probabilities[t,middle+i-1])
    end
end
```

```julia
plot([-time:time],Probabilities[time,:]')
```

Esta es una distribución gaussiana

```julia
function walk1D(time,p=0.5)
    q=1-p
    middle=time+1
    Probabilities=zeros(time,2time+1)
    Probabilities[1,middle]=1
    for t in 1:(time-1)
        for i in -t:t
            Probabilities[t+1,middle+i]=(q*Probabilities[t,middle+i+1]+p*Probabilities[t,middle+i-1])
        end
    end
    #plot([-time:time],Probabilities[time,:])
    return Probabilities'
end
```

```julia
time=1000
prob=walk1D(time,0.3)
plot([-time:time],prob[:,time])
```

## 2.

```julia
function walker2D(time,l=0.25,r=0.25,u=0.25,d=0.25)
    pz=zeros(2time+1,2time+1,time)
    middle=time+1
    pz[middle,middle,1]=1
    for t in 1:time-1
        for i in -t:t
            for j in -t:t
                    pz[middle+i,middle+j,t+1]=(l*pz[middle+i+1,middle+j,t]+r*pz[middle+i-1,middle+j,t]+d*pz[middle+i,middle+j+1,t]+u*pz[middle+i,middle+j-1,t])
            end
        end
    end
    return pz
end
```

```julia
time=10
proby=walker2D(time)
step=[-time-0.5:time+0.5]
max=maximum(proby[:,:,time])
pcolormesh(step,step,proby[:,:,time],cmap="rainbow", vmin=0, vmax=max)
xlim(-time,time)
ylim(-time,time)
colorbar()
```

# 3.

Partiendo de la ecuación

$P_{t+1}(i)=pP_{t}(i-1)+qP_{t}(i+1)$

pero entendiendo que se dan pasos de $\delta x$ en tiempos de $\delta t$, la ecuación nos queda:

$P(x,t+\delta t)=pP(x-\delta x,t)+qP(x+\delta x, t)$

Haciendo desarrollo en Taylor, tanto en x como en y, en segundo y primer orden respectivamente al rededor de $P(x,t)$, obtenemos:

$P(x,t+\delta t)=P(x,t)+\delta t P_t$, 

y 

$P(x,t+\delta t)=P(x,t)+\delta x(q-p)P_x(x,t)+(\delta x)^2P_{xx}(x,t)$. 

Que al igualar: $\delta t P_t+\delta x(p-q)P_x=(\delta x)^2P_{xx}$ . 

Dándonos como resultado:

$\frac{\partial P(x,t)}{\partial t}+ v\frac{\partial P(x,t)}{\partial x}=D\frac{\partial^2 P(x,t)}{{\partial x}^2}$

Donde definimos $v= (p-q)\frac{\delta x}{\delta t}$ y  $D=\frac{(\delta x)^2}{\delta t}$

Esta ecuación es  conocida como la ecuación de convección-difusión en una sola dimensión. Se puede ver que si la probabilidad de saltar a la derecha, $p$, y la probabilidad de saltar a la izquierda, $q$, son iguales, el segundo término desaparece y nos queda tan sólo la ecuación de difusión.

$\frac{\partial P(x,t)}{\partial t}=D\frac{\partial^2 P(x,t)}{{\partial x}^2}$

La solución a la ecuación de convección-difusión es bien conocida y es:

$P (x, t) =\frac{1}{\sqrt{4 \pi D t}} e^{−(x−vt)^2/4D}$

Para el caso simétrico, i.e. $v=0$:

$P (x, t) =\frac{1}{\sqrt{4 \pi D t}} e^{−x^2/4D}$

Que es una gaussiana y es justo lo encontrado. Ahora, algo importante es que el término $(\delta x)^2 /\delta t$ debe ser constante conforme $\delta x$ y $\delta t$ tiendan a cero.

```julia
function walker_boundaries1D(time,box_l,x,p=0.5)
    q=1-p
    Probabilities=zeros(box_l,time)
    Probabilities[x,1]=1
    for t in 1:(time-1)
        Probabilities[1,t+1]=q*(Probabilities[1,t]+Probabilities[2,t])
        Probabilities[box_l,t+1]=p*(Probabilities[box_l,t]+Probabilities[box_l-1,t])
        for i in 2:box_l-1
            Probabilities[i,t+1]=(q*Probabilities[i+1,t]+p*Probabilities[i-1,t])
        end
    end
    return Probabilities
end
```

```julia
time=1000
box_l=20
x=2
p_b=walker_boundaries1D(time,box_l,x)
plot([1:box_l],p_b[:,time])
ylim(0,1)
xlim(1,box_l)
```

```julia
time=1000
box_l=20
x=2
p_b=walker_boundaries1D(time,box_l,x,0.25)
plot([1:box_l],p_b[:,time])
ylim(0,1)
xlim(1,box_l)
```

Se espera que conforme avance el tiempo, si ambos saltos son igualmente probables, una vez alcanzadas las fronteras, se tenderá a un promedio y será igualmente probable estar en cualquier punto de la caja. Si $p\neq q$, no se alcanzará un promedio.

```julia
function walker_boundaries2D(time,box_l,x,y,r=0.25,l=0.25,u=0.25,d=0.25)
    Probabilities=zeros(box_l,box_l,time)
    Probabilities[x,y,1]=1
    for t in 1:time-1
        for i in 1:box_l
            for j in 1:box_l
                if i==1 
                    Probabilities[i,j,t+1]=l*Probabilities[i+1,j,t]+l*Probabilities[i,j,t]
                    if j==1
                        Probabilities[i,j,t+1]+=d*Probabilities[i,j+1,t]+d*Probabilities[i,j,t]
                    elseif j==box_l
                        Probabilities[i,j,t+1]+=u*Probabilities[i,j,t]+u*Probabilities[i,j-1,t]
                    else
                        Probabilities[i,j,t+1]+=d*Probabilities[i,j+1,t]+u*Probabilities[i,j-1,t]
                    end
                elseif i==box_l
                    Probabilities[i,j,t+1]=r*Probabilities[i,j,t]+r*Probabilities[i-1,j,t]
                    if j==1
                        Probabilities[i,j,t+1]+=d*Probabilities[i,j+1,t]+d*Probabilities[i,j,t]
                    elseif j==box_l
                        Probabilities[i,j,t+1]+=u*Probabilities[i,j,t]+u*Probabilities[i,j-1,t]
                    else
                        Probabilities[i,j,t+1]+=d*Probabilities[i,j+1,t]+u*Probabilities[i,j-1,t]
                    end
                else
                    Probabilities[i,j,t+1]=l*Probabilities[i+1,j,t]+r*Probabilities[i-1,j,t]
                    if j==1
                        Probabilities[i,j,t+1]+=d*Probabilities[i,j+1,t]+d*Probabilities[i,j,t]
                    elseif j==box_l
                        Probabilities[i,j,t+1]+=u*Probabilities[i,j,t]+u*Probabilities[i,j-1,t]
                    else
                        Probabilities[i,j,t+1]+=d*Probabilities[i,j+1,t]+u*Probabilities[i,j-1,t]
                    end
                end
            end
        end
    end
    return Probabilities
end
```

```julia
time=200
box_l=21
x=11
y=11
pxy_b=walker_boundaries2D(time,box_l,x,y)
max=maximum(pxy_b[:,:,time])
pcolormesh(pxy_b[:,:,time],cmap="rainbow", vmin=0, vmax=max)
xlim(1,box_l)
ylim(1,box_l)
colorbar()
```
