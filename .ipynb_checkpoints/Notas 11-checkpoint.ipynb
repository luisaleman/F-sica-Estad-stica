#FactCheck

```julia
using FactCheck
```

```julia
facts ("Ising") do
    context("Energía del Sitema") do
        σ=ones(50,50)
        E= Ising.energia(σ)
        @fact E =>-(2*(50*50))
        σ1=ones(50,50)
        E1= Ising.energia(σ1)
        @fact E1 =>-(2*(50*50))
        σ2=[[1 1], [1 -1]]
        E2= Ising.energia(σ2)
        @fact E2 =>0 
    end
    context("Diferencia de Energías") do
        σ=ones(50,50)
        for t in 1:10
            i=rand(1:50)
            j=rand(1:50)
            dE=Ising.dE(σ,i,j)
            @fact dE =>8
        end
    end
    context("Probabilidad de Aceptación") do
        σ=ones(50,50)
        i=rand(1:50)
        j=rand(1:50)
        E1=Ising.dE(σ,i,j)
        b=min(e^(-E1),1)
        #estoy haciendo T=1
        @fact b=>e^(-8)
    end
end
```
