### [1]

Tenemos para el caso de un caminante en una dimensión, con probabilidad $p$ de saltar a la derecha, $q$ a la izquierda, y $s$ de mantenerse en el mismo sitio, la siguiente ecuación maestra:

$P_{t+1}(i)=P_{t}(i)+[pP_{t}(i-1)+qP_{t}(i+1)]-[pP_{t}(i)+qP_{t}(i)]$

Donde el primer corchete es la probabilidad de que salten de los lugares adyacentes a él, y el segundo corchete es la probabilidad de que estando en el lugar, salte a los adyacentes. Haciendo un poco de álgebra, y utilizando que $p+q+s=1$, se puede llegar a:

$P_{t+1}(i)=pP_{t}(i-1)+qP_{t}(i+1)+sP_{t}(i)$

```julia
function walk1D(time,p=0.5,q=0.5,s=0.0)
    middle=time+1
    Probabilities=zeros(2time+1,time)
    Probabilities[middle,1]=1
    for t in 1:(time-1)
        for i in -t:t
            Probabilities[middle+i,t+1]=(q*Probabilities[middle+i+1,t]+p*Probabilities[middle+i-1,t]+s*Probabilities[middle+i,t])
        end
    end
    return Probabilities
end
```

```julia
time=1000
p=1./3.
q=1./3.
s=1./3.
prob=walk1D(time,p,q,s)
prob_no_s=walk1D(time)
plot([-time:time],prob_no_s[:,time])
plot([-time:time],prob[:,time])
xlim(-200,200)
xlabel("x")
ylabel("P(x)")
```

```julia
function walk1D_dif(time,Prob_i,p=0.5,q=0.5,s=0.0)
    middle=time+1
    Probabilities=zeros(2time+1,time)
    Probabilities[:,1]=copy(Prob_i)
    for t in 1:(time-1)
        for i in 2:2*time
            Probabilities[i,t+1]=(q*Probabilities[i+1,t]+p*Probabilities[i-1,t]+s*Probabilities[i,t])
        end
    end
    return Probabilities
end
```

Las siguientes dos gráficas muestran la distribución de probabilidad dado una probabilidad inicial repartida en dos puntos, y la gráfica de $t^{1/2} P_t(x)$ en función de $x/t^{1/2}$. Se sigue viendo como convergen van a converger en una función gaussiana, conforme avance el tiempo.

```julia
time=2000
middle=time+1
lim=ceil(time/60)
Prob_i=zeros(2time+1)
Prob_i[middle-lim]=0.2
Prob_i[middle+lim]=0.8
t_step=ceil(time/5)
prob=walk1D_dif(time,Prob_i)
for t in 100:t_step:time
    plot([-time:time],prob[:,t])
end
xlim(-200,200)
xlabel("x")
ylabel("P(x)")
```

Otra manera de graficar los resultados utilizando el paquete Interact:

```julia
function distribucion(tiempo::Int,S,P,Q,L)
    #Ésto nos dará el tamaño de nuestra caja    
    valores_posibles=[-L:L] #Nos genera un vector con las posiciones posibles que puede tomar nuestro caminante
    probabilidad=zeros(length(valores_posibles)) #Primer vector para hacer la distribución
    probabilidad2=zeros(length(valores_posibles)) #Segundo vector para hacer la distribución
    probabilidad[L+1]=1/3
    probabilidad[L+6]=1/3
    probabilidad[L-4]=1/3
     #Condiciones iniciales de nuestro sistema, tendremos tres puntos con probabilidad distribuida entre ellos
    #Aquí ponemos la regla de recurrencia que seguirá nuestra función para encontrar la distribución de probabilicad
    for t=1:tiempo
       for i=2:length(probabilidad)-1
            probabilidad2[i]=probabilidad[i-1]*Q+probabilidad[i+1]*P+probabilidad[i]*S
        end
        probabilidad,probabilidad2=probabilidad2,probabilidad
    end
    #La función nos regresa las probabilidades en un vector junto con las posiciones posibles en otro..
    return probabilidad, valores_posibles
end
```

```julia
fig=figure(figsize=(6,4))
@manipulate for tiempo=1:20
    withfig(fig) do
         plot(distribucion(tiempo,0.1,0.45,0.45,20)[2]/sqrt(tiempo),distribucion(tiempo,0.1,0.45,0.45,20)[1]*sqrt(tiempo),"r")
         xlabel("x/t^(1/2)")
        ylabel("P_t(x)*t^(1/2)")
        title("Distribución de probabilidad")
        end
end
```

## Tiempos de primer paso: Sistemas con absorción

Ahora usaremos el método de enumeración para un sistema con paredes y además con un pozo en cero.

#### Implementación

Haremos una función que cumpla con las características:
   - Si va a saltar fuera de la caja con cierta probabilidad, dicha probabilidad se sumará a la probabilidad de que se quede en la frontera en vez de eso.
   - Si llega al cero con cierta probabilidad, dicha probabilidad se guarda en un valor aparte.

#### Función: distribucion2(tiempo,R,Q,L)

1. Entradas
    - tiempo: corresponde al tiempo en el que tendremos expresada nuestra distribución de salida.
    - R: es la probabilidad de que nuestro caminante se quede en el mismo sitio.
    - Q: es la probabilidad de que cuando de un paso, éste sea hacia la derecha.
    - L: es el tamaño de la caja
2. Salidas
    - distribucion2[1]: nos arroja un vector con las probabilidades para cada posición 
    - distribucion2[2]: nos da la probabilidad de que el caminante ya se haya caído al pozo

```julia
function distribucion2(tiempo::Int,R,Q,L)    
    valores_posibles=[0:L] #Nos genera un vector con las posiciones posibles que puede tomar nuestro caminante
    probabilidad=zeros(length(valores_posibles)) #Primer vector para hacer la distribución
    probabilidad2=zeros(length(valores_posibles)) #Segundo vector para hacer la distribución
    probabilidad3=0.0
    probabilidad[2]=1   #Condiciones iniciales de nuestro sistema
    frontera=length(probabilidad)
  
    #Aquí ponemos la regla de recurrencia que seguirá nuestra función para encontrar la distribución de probabilicad
    for t=1:tiempo
       probabilidad3+=probabilidad[2]*(1-R)*(1-Q) #Probabilidad de caer en el pozo y desaparecer
       probabilidad2[frontera]=probabilidad[frontera]*(R+(1-R)*Q)+probabilidad[frontera-1]*(1-R)*Q 
       for i=2:frontera-1
            probabilidad2[i]=probabilidad[i-1]*(1-R)*(1-Q)+probabilidad[i+1]*(1-R)*Q+probabilidad[i]*R
        end
        probabilidad,probabilidad2=probabilidad2,probabilidad
    end
    #La función nos regresa las probabilidades en un vector junto con las posiciones posibles en otro..
    return probabilidad,probabilidad3
    
end
```

Para asegurar que se conserva la probabilidad, deberíamos observar que la probabilidad de que caiga sumada a las probabilidades de estar en la misma posición deben ser iguales a 1. Ésto lo podemos ver a continuación:

```julia
sum(distribucion2(1000,0.5,0.5,50)[1])+distribucion2(1000,0.5,0.5,50)[2]
```

Haremos una nueva función para guardar las probabilidades de caerse en el pozo en un tiempo $T$ y para distintos valores de $L$

#### Función: promedio_cajas(T,L)

1. Entradas
    - T: es el tiempo que habrá pasado
    - L: es el tamaño de la caja máxima que tendremos
2. Salidas
    - promedio_cajas: es la probabilidad de que caiga en el pozo al tiempo T en las distintas cajas

```julia
function promedio_cajas(T,L)
    probabilidad=Float64[];
    for l=1:L
        push!(probabilidad,distribucion2(T,0.5,0.5,l)[2])
    end
    return probabilidad
end
```

#### Función 5: distribucion_tiempo(T,L)

1. Entradas
    - T: es el tiempo máximo que graficaremos
    - L: es el tamaño de la caja máxima que tendremos
2. Salidas
    - Obtendremos varias gráficas para distintos tiempos $T$ en función del tamaño de la caja 

```julia
function distribucion_tiempo(T,L)
    for t=1:T
    plot(1:L,promedio_cajas(t,L))
    end
    title("Distribución de probabilidad para distintos T en distintas cajas")
    xlabel("Tamaño de caja")
    ylabel("Probabilidad")
end
```

```julia
distribucion_tiempo(10,5)
```

Las siguientes funciones realizan una distribución de la probabilidad de los distintos tiempos de llegada al cero, dependiendo el tamaño de la caja, similar a lo realizado en la taraea pasada.

```julia
function walker_boundaries1D_simple(time,box_l,x,p=0.5,q=0.5,s=0)
    Probabilities=zeros(box_l,time)
    Probabilities[x,1]=1
    for t in 1:(time-1)
        Probabilities[1,t+1]=q*(Probabilities[2,t])+s*Probabilities[1,t]
        Probabilities[box_l,t+1]=p*(Probabilities[box_l,t]+Probabilities[box_l-1,t])+s*Probabilities[box_l,t]
        for i in 2:box_l-1
            Probabilities[i,t+1]=(q*Probabilities[i+1,t]+p*Probabilities[i-1,t])+s*Probabilities[i,t]
        end
    end
    return Probabilities
end
```

```julia
function walker_boundaries1D(time,box_l,x,p=0.5,q=0.5,s=0)
    Probabilities=zeros(box_l,time)
    Probabilities[x,1]=1
    P_o=[]
    for t in 1:(time-1)
        if t>x-1
            push!(P_o,q*Probabilities[1,t]) 
        end
        Probabilities[1,t+1]=q*(Probabilities[2,t])+s*Probabilities[1,t]
        Probabilities[box_l,t+1]=p*(Probabilities[box_l,t]+Probabilities[box_l-1,t])+s*Probabilities[box_l,t]
        for i in 2:box_l-1
            Probabilities[i,t+1]=(q*Probabilities[i+1,t]+p*Probabilities[i-1,t])+s*Probabilities[i,t]
        end
        #Probabilities/=sum(Probabilities[:,t])
    end
    return Probabilities, P_o
end
```

```julia
time=1000
box_l=10
x=5
p=1./3.
q=1./3.
s=1./3.
prob_box=walker_boundaries1D_simple(time,box_l,x,p,q,s)
t_step=ceil(time/5)
x=[1:box_l]
for t in 10:t_step:time
    plot(x,prob_box[:,t])
end
```

En la gráfica de arriba se muestra que si en un borde los caminantes desaparecen, la probabilidad no se conservaría, y tendería a 0, si no se escoge alguna regla de reemplazo del "caminante".
