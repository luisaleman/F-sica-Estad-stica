```julia
function generate(N) 
    N1=0
    N2=0
    N3=0
    N4=0
    N5=0
    N6=0
    for i in 1:N
        x=rand(1:6)
        if x==1
            N1+=1
        elseif x==2
            N2+=1
        elseif x==3
            N3+=1
        elseif x==4
            N4+=1
        elseif x==5
            N5+=1
        elseif x==6
            N6+=1
        end                    
    end
    P1=N1/N
    P2=N2/N
    P3=N3/N
    P4=N4/N
    P5=N5/N
    P6=N6/N
    plot(1,P1,"o")
    plot(2,P2,"o")
    plot(3,P3,"o")
    plot(4,P4,"o")
    plot(5,P5,"o")
    plot(6,P6,"o")
end
```

```julia
function generatenew(N,Ni,Pi) 
    var=0
    faces=int(length(Ni))
    for i in 1:N
        x=rand(1:faces)
        for j in 1:faces
            if x==j
                Ni[j]+=1
            end
        end
    end
    for i in 1:faces
        Pi[i]= Ni[i]/N
    end
    
end
```

```julia
function bernoulli(throws,P,exitos)
    M=0
    for i in 1:throws
        if rand()<=P
            M+=1
        end  
    end
    exitos[M+1]+=1
end
```

```julia
N=6
pseudoexperiments=10
throws=[0:N]
exitos=zeros(N+1)
for i in 1:pseudoexperiments
    H=bernoulli(N,1./6.,exitos)
end
    exitos/=(pseudoexperiments*N)
    plot(throws,exitos,"o")
```
