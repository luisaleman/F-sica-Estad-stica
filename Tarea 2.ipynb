```julia
function Walker(L)
    position=1
    time=0
    while position>0
        position+=rand([-1,1])
        if position>L
            position=L
        end
        time+=1
    end
    return time
end
```

[2]

```julia
function Average(N_runs,L)
    τ=0.0
    for i in 1:N_runs
        τ+=Walker(L) 
    end
    τ/=N_runs
    return τ
end
```

[3]

```julia
N_runs=10000
L=zeros(4)
time=zeros(4)
for i in 1:4
    L[i]=10^(i-1)
    time[i]=Average(N_runs,L[i])
    println(time[i])
end
```

```julia
plot(L,time)
xlabel("L")
ylabel("<t>")
```

El tiempo crece linealmente. Cuando L se vaya a infinito, si el número de corridas "N_run" no es lo suficientemente grande, dejará de crecer exponencialmente y tendrá un valor finito (véase la siguiente gráfica que muestra lo que sucede cuando el número de corridas no es suficiente). Sin embargo, si el número de corridas también tiende a infinito, el tiempo promedio de igual forma lo hará.

```julia
N_runs=10
L_w=zeros(8)
time_w=zeros(8)
for i in 1:8
    L_w[i]=10*i
    time_w[i]=Average(N_runs,L_w[i])
end
```

```julia
plot(L_w,time_w)
xlabel("L")
ylabel("<t>")
```

Aquí se muestra que el número de corridas no puede ser mucho menor a L

```julia
function Probability_dist(N_runs,L)
    times=zeros(Int,N_runs)
    for i in 1:N_runs
        times[i]=Walker(L)
    end
    t_max=maximum(times)
    distribution=zeros(t_max)
    for t in 1:t_max
        for i in 1:N_runs
            if t==times[i]
                distribution[t]+=1
            end
        end
    end
    distribution/=N_runs
    return distribution
end
```

```julia
N=100000
dist=Probability_dist(N,2)
plot(dist,"o")
yscale("log")
```

Esta es la distribución de probabilidad de una sola $\tau_L$

```julia
N=100000
dist=Probability_dist(N,4)
plot(dist,"o")
yscale("log")
```

```julia
N=10000
num=7
for i in 1:num
    dist=Probability_dist(N,2*i)
    plot(dist,label=2*i,"o",markersize=num+4-i)
    yscale("log")
    ylim(0.0001,1)
    xlim(0,150)
    legend()
end
```

```julia
N=10000
num=7
for i in 1:num
    dist=Probability_dist(N,2*i)
    plot(dist,label=2*i,"o",markersize=num+4-i)
    yscale("log")
    ylim(0.0001,1)
    xlim(0,500)
    legend()
end
```
