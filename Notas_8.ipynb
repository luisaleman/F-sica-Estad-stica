# Física estadística en el ensamble canónico 

Ya que contamos con un modelo microscópico, el modelo de Ising, que especifica cuáles son las *configuraciones microscópicas*, o **microestados**, del sistema, y su energía, ¡podemos hacer **física estadística**!

Por el momento, y en la mayor parte del curso, ocuparemos el **ensamble canónico**. Recordemos que esto modela un sistema en contacto con un baño térmico (o reservorio) infinito, a una temperatura fija $T$. 

[Estos conceptos no son nada sencillos, pero supondremos que ya se han tratado en el curso de Física Estadística.]

$\newcommand{\ss}{\pmb{\sigma}}$
Lo que debemos extraer de esto es que un sistema en equilibrio, a temperatura $T$, tiene cierta probabilidad de ocupar cada microestado $\ss$, dada por

$$p_T^{\mathrm{(eq)}}(\ss) \propto e^{-\beta E(\ss)},$$

donde hemos introducido la **temperatura inversa** $\beta := 1/(k_B T)$, con $k_B$ la constante de Boltzmann. La cantidad $\beta$ resulta ser más natural que pensar en términos de $T$; tomaremos unidades en las cuales $k_B = 1$.

Para la constante de proporcionalidad, introducimos la notación $Z(\beta)$, que depende de la temperatura o de $\beta$, de tal forma que las probabilidades están dadas por

$$p_T^{\mathrm{(eq)}}(\ss) = \frac{1}{Z(\beta)} e^{-\beta E(\ss)}.$$


Esta distribución de probabilidad se llama la **distribución canónica** a temperatura $T$, o la **distribución de Boltzmann(-Gibbs)**.

[1] Escribe una expresión para $Z(\beta)$, que se llama la **función de partición** a temperatura $T$. Recuerda cuál es la relación entre $Z$ y la energía libre macroscópica, $F$.

La función de partición, $Z(\beta)$se puede obtener de la siguiente manera, dada la definición que se nos da anteriormente 
\begin{equation}
1=\sum_s p_T(s)=\sum_s\frac{1}{Z}e^{-\beta \epsilon_s}=\frac{1}{Z}\sum_s e^{-\beta \epsilon_s}
\end{equation}
donde $\epsilon_s$ es la energía asociada al estado $s$. Ahora, despejando $Z$
\begin{equation}
Z(\beta)=\sum_s e^{-\beta \epsilon_s}.
\end{equation}
En mecánica clásica, ya que se cuenta con un continuo de posibles estados, la función de partición quedaría para N partículas,
\begin{equation}
Z=\frac{1}{N!h^{3N}}\int e^{-\beta H(\bar{q},\bar{p})}dQdP
\end{equation}
La relación entre la energía libre de Helmholtz y $Z$ es
\begin{equation}
F=-k_BT\ln(Z)
\end{equation}

Las cantidades de interés son cantidades macroscópicas, dadas por **promedios** sobre la distribución de Boltzmann, que denotaremos con $\langle \cdot \rangle$. En general, estas propiedades se pueden expresar en términos de la función de partición, $Z$, por lo cual $Z$ asume un papel principal en la física estadística de equilibrio.

[2] Escribe una expresión para el promedio $\langle E \rangle_\beta$ de la energía a temperatura $T$. Demuestra que se puede expresar en términos de una derivada de $Z(\beta)$.

Tenemos que la energía promedio está dada por la energía del posible sistema $s$ multiplicada por la probabilidad de dicho sistema, i.e.
\begin{equation}
\langle E \rangle_\beta=\sum_s \epsilon_s P_s=\sum_s \epsilon_s \frac{1}{Z}e^{-\beta \epsilon_s}=\frac{1}{Z}\sum_s \epsilon_s e^{-\beta \epsilon_s}
\end{equation}
dónde podemos ver la última expresión como una derivada
\begin{equation}
\frac{1}{Z}\sum_s \epsilon_s e^{-\beta \epsilon_s}=-\frac{1}{Z}\frac{\partial}{\partial \beta}\sum_s e^{-\beta \epsilon_s}=-\frac{1}{Z}\frac{\partial Z}{\partial \beta}=-\frac{\partial \ln Z}{\partial \beta}
\end{equation}
por lo tanto
\begin{equation}
\langle E \rangle_\beta=-\frac{\partial \ln Z}{\partial \beta}
\end{equation}

En principio, podemos calcular $Z$, y los promedios, de forma exacta a partir de su definición, como sigue.

[3] (i) Escribe código que genera *todos los microestados $\ss$* del modelo de Ising cuadrado de tamaño $n \times m$, uno tras otro. [Pista: un arreglo 2D en Julia se puede manipular como si fuera de 1D; piensa en binario.]

(ii) Utiliza tu código para ir calculando $Z(\beta)$ y $\langle E \rangle$ para distintas temperaturas *para sistemas muy pequeños*. Utiliza condiciones periódicas de frontera cuando calculas $E(\ss)$. [Empieza con el sistema más pequeño y aumenta el tamaño de uno en uno...]


```julia
function microestados(n,m,T=1.0)
    n_spins=n*m
    top=2^n_spins
    Z=0.0
    E=0.0
    mag=0.0
    energy=0.0
    for j in 1:top
        M_seed=bin(j-1,n_spins)
        M=zeros(Int,n_spins)
        for i in 1:n_spins
            M[i]=parseint(M_seed[i])*2-1
        end
        #println(M)
        for i in 1:n_spins
            energy-=M[i]*M[fld(i-1,m)*m+mod1(i+1,m)]
            energy-=M[i]*M[mod1(i+m,n_spins)]
        end
        Z+=exp(-energy/T)
        E+=energy*exp(-energy/T)
        mag=sum(M)*exp(-energy/T)
        #println(energy)
        energy=0.0    
    end
    E/=Z
    mag/=Z
    return Z,E,mag
end
```

```julia
@time microestados(5,5)
```

```julia
@time microestados(5,5,20)
```

Hasta un tamaño de $5\times 5$, más allá se tarda demasiado. Ya que el tiempo va como $2^{m\times n}$, es decir, de un sistema de $5\times 5$ a uno de $4\times5$, se hace $2^5=32$ veces más tiempo.

(iv) Para el sistema más grande que puedas, dibuja estas cantidades en función de $T$. ¿Qué observas?


```julia
function Z_E_plot(n,m,T_max,n_points, log=false)
    step=T_max/n_points
    Z=zeros(n_points)
    E=zeros(n_points)
    mag=zeros(n_points)
    T=zeros(n_points)
    for i in 1:n_points
        Z[i],E[i],mag[i]=microestados(n,m,step*i)
        T[i]=step*i
    end
    plt.subplot(311)
    plot(T,Z)
    plt.ylabel("Z")
    if log==true
        plt.yscale("log")
    end
    plt.subplot(312)
    plot(T,E)
    plt.ylabel("E")
    plt.subplot(313)
    plot(T,mag)
    plt.ylabel("M")
    plt.xlabel("T")
    if log==true
        println(mag)
    end
end
```

[4] Hay dos valores especiales de la temperatura. 

(i) ¿Cuáles son? 

Pues son los $T <1$ y $T>1$ (en nuestras unidades especiales). Para $T >>1$ la magnetización, $M$, decae de manera exponencial, mientras que la energía, $E$, aumenta y tiende a $0$ asintóticamente. Para $T<<1$,$M$  empieza en un valor fijo, decayendo lentamente (es casi constante antes de $T=1$), y  $E$ comienza a crecer lentamente (de igual forma, es casi constante antes de $T=1$).

(ii) Para estas temperaturas especiales, calcula analíticamente $Z$ $\langle E \rangle$. ¿Qué debería pasar entre estas dos temperaturas?

Para $T>>1$, tenemos que $\beta\approx0$, por lo que 

\begin{equation}
Z(\beta)=\sum_s e^{-\beta \epsilon_s}\approx=\sum_s 1= N_s
\end{equation}

dónde $N_s$ es el número de posibles estados del sistema. De esta forma, la energía $\langle E \rangle$ se transforma
en un promedio de todas las energías de los posibles estados. Por  nuestro sistema, este promedio será justamente 0 (porque hay la misma cantidad de estados con energía "positiva" como "negativa". De igual forma, la magnetización $\langle M \rangle$ será el promedio, i.e. cero.

Para $T<<1$, tenemos que $\beta>>1$, por lo que habrá problemas con $Z$ por los términos que tengan $\epsilon<0$ (que dominarán) y esta comenzará a crecer, y en el caso límite $T\rightarrow 0$, $Z\rightarrow \infty$.  La energía  por ende será "negativa", ya que sus términos son lo que dominan. 

Físicamente, lo que se interpreta en el primer caso, es que cuando $T>>1$ (en nuestras unidades), se vuelven igual de probables todos los estados o configuraciones, sin importar su energía, y por ello, los promedios de las medidas, energía y magnetización, de vuelven promedios y tienden a cero.

Entre estas dos temperaturas hay una transición de fase, ya que pasa de un estado magnetizado, a uno que no, y tanto el comportamiento de la energía como la magnetización lo demuestran. Son cambios que no son discontinuos a primer orden pero sí a segundo.

## ¿Qué sigue? 

En principio, si calculamos $Z$ como función de $\beta$, podemos extraer las propiedades termodinámicas macroscópicas. Pero acabamos de ver que ¡*no* podemos calcular $Z$ para sistemas de tamaño interesante!

¿Cuál es la solución? Debemos encontrar una manera de calcular las cantidades de interés, como $\langle E \rangle$, *sin* tener que pasar por la vía de calcular $Z$.

Dado que estas cantidades tienen que ver con *promedios*, surge la idea de calcularlas al promediar sobre *datos* que generamos. ¿Cuáles datos? La idea es que generemos una secuencia *en el tiempo* de configuraciones $\ss^{(1)}$, $\ss^{(2)}$, etc., y que calculemos $\langle E \rangle$ como el promedio aritmético --en el sentido de la estadística-- de *los datos* $E(\ss^{(1)})$, $E(\ss^{(2)})$, etc. Por supuesto, esto nos dará sólo *una aproximación* al valor exacto. Si la generación de la secuencia ocupa un proceso azaraso, entonces *tendremos un método tipo Monte Carlo*.
